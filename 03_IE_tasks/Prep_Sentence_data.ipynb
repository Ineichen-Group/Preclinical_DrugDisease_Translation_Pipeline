```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('./data/annotated/full_ner/all_annotations_minimal_fixed_multi.csv')
df['tokens'] = df['tokens'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df['ner_tags'] = df['ner_tags'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df['annotations_tuples'] = df['annotations_tuples'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python

# Create a list to hold the new rows
rows = []

for _, row in df.iterrows():
    doc_id = row['doc_id']
    
    for s_idx, e_idx, ent_type, ent_text in row['annotations_tuples']:
        if ent_type in ("age", "weight"):
            new_row = {
                'doc_id': doc_id,
                'ent_type': ent_type,
                'ent_text': ent_text,
                's_idx': s_idx,
                'e_idx': e_idx
            }
            rows.append(new_row)

# Create the new DataFrame
new_df = pd.DataFrame(rows)
new_df['doc_id_unique'] = new_df['doc_id'].astype(str) + "_" + new_df['s_idx'].astype(str) + "_" + new_df['e_idx'].astype(str)
```

```python
new_df
```

```python
# Get unique doc_id_unique per entity type
age_docs = set(new_df[new_df['ent_type'] == 'age']['doc_id_unique'].unique())
weight_docs = set(new_df[new_df['ent_type'] == 'weight']['doc_id_unique'].unique())

# Intersect to find docs with both
both_docs = age_docs & weight_docs

# Print results
print("Unique with age:", len(age_docs))
print("Unique with weight:", len(weight_docs))
print("Unique with both:", len(both_docs))
```
