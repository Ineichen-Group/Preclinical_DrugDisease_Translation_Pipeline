```python
from collections import defaultdict

unique_drugs_with_mappings = {}        # drug -> first seen mapping
drug_frequency = defaultdict(int)      # drug -> count
mapped_to_ner = defaultdict(set)      # "drug(mapping)" -> list of NER originals

for _, row in preclin_mapped_drug.iterrows():
    drug_names    = str(row['drug_term_umls_norm']).split("|")
    drug_mappings = str(row['drug_umls_termid']).split("|")
    original_ner  = str(row['linkbert_mapped_drugs']).split("|")

    # Warn if names vs mappings differ
    if len(drug_names) != len(drug_mappings):
        print(f"ERROR (names vs mappings length) PMID {row['PMID']}")
        print(f"  drug_names   : {drug_names}")
        print(f"  drug_mappings: {drug_mappings}")
        continue

    # We'll iterate up to the shortest to avoid IndexErrors
    min_len = min(len(drug_names), len(drug_mappings))

    for i in range(min_len):
        drug = drug_names[i].strip()
        mapping = drug_mappings[i].strip()
        if not drug:
            continue

        # record mapping only once (first seen wins)
        unique_drugs_with_mappings.setdefault(drug, mapping)

        # count frequency
        drug_frequency[drug] += 1

        # align to original_ner if available at the same position
        if len(drug_names) != len(original_ner):
            print(f"NOTE (names vs NER length) PMID {row['PMID']}: "
                  f"{len(drug_names), drug_names} names vs {len(original_ner), original_ner} NER tokens")
            
        elif i < len(original_ner) and mapping!="-1":
            ner_token = original_ner[i]
            key = f"{drug} ({mapping})"
            mapped_to_ner[key].add(ner_token)

    
```

```python
mapped_to_ner_df = (
    pd.DataFrame([
        {
            "drug_mapping": key,
            "drug": key.split("(")[0],
            "mapping": key.split("(")[1].rstrip(")"),
            "ner_tokens": values,
            "ner_count": len(values)
        }
        for key, values in mapped_to_ner.items()
    ])
)

# Optional: sort by count descending
mapped_to_ner_df = mapped_to_ner_df.sort_values(by="ner_count", ascending=False).reset_index(drop=True)
mapped_to_ner_df
```

```python
# Total unique drugs
count_unique_entities = len(unique_drugs_with_mappings)

# Unique drugs that mapped successfully
mapped_unique_tokens = sum(1 for m in unique_drugs_with_mappings.values() if m != '-1')

# % mapped unique success
pct_mapped_unique_success = (mapped_unique_tokens / count_unique_entities) * 100 if count_unique_entities > 0 else 0

# % mapped frequency-adjusted
total_mentions = sum(drug_frequency.values())
mapped_mentions = sum(freq for drug, freq in drug_frequency.items()
                      if unique_drugs_with_mappings.get(drug) != '-1')
pct_mapped_freq_adjusted = (mapped_mentions / total_mentions) * 100 if total_mentions > 0 else 0

# Unique success tells you coverage — how many different drugs you can map.
# Frequency-adjusted success tells you impact — how many mentions in your dataset are successfully linked, which matters if a few drugs appear very frequently.
# --- Summary Table ---
summary = pd.DataFrame([{
    "count unique entities": count_unique_entities,
    "mapped_unique_tokens": mapped_unique_tokens,
    "% mapped unique success": round(pct_mapped_unique_success, 2),
    "% mapped frq adjusted": round(pct_mapped_freq_adjusted, 2)
}])

summary
```

```python
drug_df = pd.DataFrame([
    {
        "drug": drug,
        "mapping": unique_drugs_with_mappings.get(drug, None),
        "frequency": drug_frequency.get(drug, 0)
    }
    for drug in unique_drugs_with_mappings
])
drug_df = drug_df.sort_values(by="frequency", ascending=False).reset_index(drop=True)
```

```python
# Add helper column
drug_df['is_mapped'] = drug_df['mapping'] != '-1'

# Split into mapped and unmapped groups
mapped_top10 = (
    drug_df[drug_df['is_mapped']]
    .sort_values(by='frequency', ascending=False)
    .head(10)
    .reset_index(drop=True)
)

unmapped_top10 = (
    drug_df[~drug_df['is_mapped']]
    .sort_values(by='frequency', ascending=False)
    .head(10)
    .reset_index(drop=True)
)

# Combine side by side
side_by_side = pd.concat(
    [mapped_top10[['drug', 'frequency']].rename(columns={'drug': 'Mapped Drug', 'frequency': 'Mapped Freq'}),
     unmapped_top10[['drug', 'frequency']].rename(columns={'drug': 'Unmapped Drug', 'frequency': 'Unmapped Freq'})],
    axis=1
)
side_by_side
```
