```python
import pandas as pd
from pathlib import Path
import re
```

# Drug

```python
CHUNKS_DIR = Path(
    "/shares/animalwelfare.crs.uzh/Preclinical_Pipeline/04_normalization/data/"
    "mapped_to_embeddings_ontologies/preclinical_chunks/"
)

OUTPUT_FILE = Path(
    "/shares/animalwelfare.crs.uzh/Preclinical_Pipeline/04_normalization/data/"
    "mapped_to_embeddings_ontologies/drug_mapped_preclinical_enriched_all.csv"
)

FILENAME_REGEX = re.compile(r"drug_mapped_preclinical_drug_enriched_(\d+)\.csv")


chunk_files = []
for f in CHUNKS_DIR.iterdir():
    match = FILENAME_REGEX.match(f.name)
    if match:
        chunk_id = int(match.group(1))
        chunk_files.append((chunk_id, f))

if not chunk_files:
    raise FileNotFoundError(f"No matching chunk files found in {CHUNKS_DIR}")

# sort numerically by chunk id
chunk_files.sort(key=lambda x: x[0])

print(f"Found {len(chunk_files)} chunk files")

dfs = []
for chunk_id, f in chunk_files:
    print(f"Reading chunk {chunk_id}: {f.name}")
    df = pd.read_csv(f)
    #df["source_chunk"] = chunk_id  # optional provenance
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)

print(f"Combined shape: {combined_df.shape}")
```

```python
combined_df.head()
```

```python
combined_df.to_csv(OUTPUT_FILE, index=False)

print(f"Saved combined CSV to: {OUTPUT_FILE}")

```
