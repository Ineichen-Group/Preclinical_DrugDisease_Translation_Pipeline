```python
import re

def split_query_max_or(query, max_or=10):
    # Tokenize query into terms separated by OR
    tokens = re.split(r'\s+OR\s+', query)
    
    subqueries = []
    current_subquery = []
    or_count = 0

    for token in tokens:
        # Add token to the current subquery
        current_subquery.append(token)
        or_count += 1

        # If max OR count reached, finalize the subquery
        if or_count == max_or:
            subqueries.append(" OR ".join(current_subquery))
            current_subquery = []
            or_count = 0

    # Add the last subquery
    if current_subquery:
        subqueries.append(" OR ".join(current_subquery))

    return subqueries

```

```python
def split_query_max_or_with_parentheses(query, max_or=10):
    # Tokenize while preserving parentheses and logical operators
    tokens = re.findall(r'\(|\)|\w+\*?|\w+\#?\d*|\w+ adj\d+ \w+|AND|OR', query)
    
    subqueries = []
    current_subquery = []
    or_count = 0
    open_parentheses = 0

    for token in tokens:
        current_subquery.append(token)
        
        # Track parentheses balance
        if token == "(":
            open_parentheses += 1
        elif token == ")":
            open_parentheses -= 1
        
        # Count ORs only when not inside parentheses
        if token == "OR" and open_parentheses == 0:
            or_count += 1

        # If max OR count is reached and parentheses are balanced, finalize the subquery
        if or_count == max_or and open_parentheses == 0:
             # Ensure subquery doesn't end with OR
            if current_subquery[-1].upper() == "OR":
                current_subquery.pop()
            subqueries.append(" ".join(current_subquery))
            current_subquery = []
            or_count = 0

    # Add the last subquery
    if current_subquery:
        if current_subquery[-1].upper() == "OR":
            current_subquery.pop()
        subqueries.append(" ".join(current_subquery))

    # Check for unbalanced parentheses (optional safeguard)
    #if open_parentheses != 0:
     #   raise ValueError("Unbalanced parentheses in the query.")

    return subqueries
```
